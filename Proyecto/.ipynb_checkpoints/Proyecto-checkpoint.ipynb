### Datos no numericos

Como no todos los datos son numericos se procede a codificarla para convertirlos en num√©ricos.

Primero se trabaja con el feature 'edjefe', el cual describe los a√±os de escolaridad del hombre que encargado del hogar, donde los datos representan yes=1 y no=0

```python
data['edjefe'].head(10)
```

Se procede a codificar los datos a numericos

```python
data['edjefe'].replace({'yes':1, 'no':0}, inplace=True)
data['edjefe'] = pd.to_numeric(data['edjefe'])
```

```python
print(data['edjefe'].head(10))
data['edjefe'].dtypes
```

Ahora se aplica el mismo proceso para el feature 'edjefa', el cual en este caso representa lo mismo que el feature anterior pero aplicado para la mujer encargada del hogar.

```python
data['edjefa'].head(10)
```

```python
data['edjefa'].replace({'yes':1, 'no':0}, inplace=True)
data['edjefa'] = pd.to_numeric(data['edjefa'])
```

```python
print(data['edjefa'].head(10))
data['edjefa'].dtypes
```

Se analiza el feature 'dependency', el cual es la division entre la cantidad de integrantes del hogar menores de 19 y mayores de 65 dividido entre los integrantes comprendido entre las edades mencionadas.

```python
data['dependency'].head(5)
```

Al ser un feature donde no se especifica que representa 'yes' y 'no', se procede a descartar el feature.

```python
data.drop(columns=['dependency'], inplace=True)
```

Tampoco se va a utilizar la columna que contiene el identificador de hogar ni de individuo ya que no se considera relevante

```python
data.drop(columns=['idhogar', 'Id'], inplace=True)
```

## 2. Aplicar Random Forest como modelo de aprendizaje

Se va a aplicar el modelo de ensamble Random Forest con todos los features del set de datos procesado

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

Se divide el set de datos de entrenamiento en labels y features

```python
X_train = data.iloc[:, 0:-1]
X_train.columns
```

```python
y_train = data.iloc[:, -1]
y_train.name
```

### Preprocesado del set de prueba

Ahora se debe importar el set de datos de prueba y hacerle el mismo preprocesado de datos que son el set de entrenamiento

```python
test_data = pd.read_csv('https://raw.githubusercontent.com/luismoreracla/DataScienceTEC/master/Proyecto/test.csv')
output = pd.read_csv('https://raw.githubusercontent.com/luismoreracla/DataScienceTEC/master/Proyecto/sample_submission.csv')
```

```python
output.head(10)
```

```python
test = pd.concat([test_data, output['Target']], axis=1)
```

```python
test.drop(columns=['Id', 'idhogar'], inplace =True)
```

```python
test.columns[test.isnull().any()].tolist()
```

```python
test[['v2a1', 'v18q1', 'rez_esc', 'meaneduc', 'SQBmeaned']].isna().sum()
```

```python
test.drop(columns=['v2a1', 'v18q1', 'rez_esc'], inplace=True)
test.shape
```

```python
test.dropna(inplace=True)
test.shape
```

```python
#test.drop(columns=['edjefe', 'edjefa'], inplace=True)
```

```python
test.drop(columns='dependency', inplace=True)
```

```python
test['edjefe']
```

```python
test.head(10)
```

```python
test['edjefe'].replace({'yes':1, 'no':0}, inplace=True)
test['edjefe'] = pd.to_numeric(test['edjefe'])
```

```python
test['edjefa'].replace({'yes':1, 'no':0}, inplace=True)
test['edjefa'] = pd.to_numeric(test['edjefa'])
```

```python
test.isna().sum()
```

```python
test.columns
```

### Aplicacion del modelo de entrenamiento

```python
X_test = test.iloc[:,0:-1]
X_test.columns
```

```python
y_test = test.iloc[:,-1]
y_test.name
```

```python
random_forest = RandomForestClassifier(n_estimators=1500, random_state=1)
random_forest.fit(X_train, y_train)
```

```python
y_pred = random_forest.predict(X_test)
```

## 3. Efectividad del modelo

```python
from sklearn import metrics
```

```python
print('Accuracy', metrics.accuracy_score(y_test, y_pred))
```

## 4. Buscando los features mas relevantes del modelo

```python
imp_features = pd.Series(random_forest.feature_importances_, index=X_train.columns).sort_values(ascending=False).head(70)
imp_features
```

```python
plt.figure(figsize=(25, 25))
sns.barplot(x=imp_features, y=imp_features.index)
#sns.set(rc={'figure.figsize':(5000,5000)})
plt.xlabel('Importancia en el modelo')
plt.ylabel('Feature')
plt.title('Features mas relevantes en el modelo')
plt.legend()
#plt.figure(figsize=(100,100))
plt.show()
```
